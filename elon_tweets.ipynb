```python
tweets = EM_tweets[['Timestamp','text','Likes','Retweets']]
tweets
```

```python
#Lets clean the tweets (remove @'s' , urls and calues of comments in the end)
TEXT_CLEANING_RE = "@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+"
new = []
for t in tweets.text:
    words = t.split()
    t = t.split()
    t = ' '.join(word for word in t if not word.startswith('@'))
    t = t.split()
    t = ' '.join(word for word in t if not word.startswith('http'))
    t = t.split()[:-3]
    t = ' '.join(t)
    t = re.sub('Replying to ','',t)
    new.append(t)
new
tweets['tweet'] = new
tweets
```

```python
def clean_retweets(row):
    if 'K' in row['Retweets']:
        return int(float(row['Retweets'].split('K')[0])*1000)
    elif 'M' in row['Retweets']:
        return int(float(row['Retweets'].split('M')[0])*1000000)
    elif ',' in row['Retweets']:
        return int(float(re.sub(',','',row['Retweets'])))
    else:
        return int(row['Retweets'])
        
tweets['Retweets'] = tweets.apply(clean_retweets, axis = 1)
tweets['Retweets'] = tweets['Retweets'].astype(int)
```

```python
def clean_likes(row):
    if 'K' in str(row['Likes']):
        return int(float(row['Likes'].split('K')[0])*1000)
    elif 'M' in str(row['Likes']):
        return int(float(row['Likes'].split('M')[0])*1000000)
    elif ',' in str(row['Likes']):
        return int(float(re.sub(',','',row['Likes'])))
    elif 'nan' in str(row['Likes']):
        return 0
    else:
        return row['Likes']
    
tweets['Likes'] = tweets.apply(clean_likes, axis = 1)
tweets['Likes'] = tweets['Likes'].astype(int)
```

```python
tweets.dtypes
```

```python
tweets.drop(['text'], axis = 1, inplace =True)
tweets.columns = ['Date', 'Likes', 'Retweets', 'Tweet']
tweets
```

```python
#Now dogecoin
doge = pd.DataFrame(columns = ['Date', 'Tweet', 'Likes', 'Retweets'])

for i in range(len(tweets)):
    cur = tweets.loc[i].Tweet
    cur = cur.lower()
    if 'doge' in cur:
        doge = doge.append({'Date' : tweets.loc[i].Date, 'Tweet' : tweets.loc[i].Tweet, 'Likes' : tweets.loc[i].Likes, 'Retweets' : tweets.loc[i].Retweets}, 
                ignore_index = True)

doge['Date'] = pd.to_datetime(doge['Date'], format='%Y-%m-%d')
doge
```

```python
#Now bitocin
btc = pd.DataFrame(columns = ['Date', 'Tweet', 'Likes', 'Retweets'])

for i in range(len(tweets)):
    cur = tweets.loc[i].Tweet
    cur = cur.lower()
    if ('btc' in cur) | ('bitcoin' in cur):
        btc = btc.append({'Date' : tweets.loc[i].Date, 'Tweet' : tweets.loc[i].Tweet, 'Likes' : tweets.loc[i].Likes, 'Retweets' : tweets.loc[i].Retweets}, 
                ignore_index = True)

btc['Date'] = pd.to_datetime(btc['Date'], format='%Y-%m-%d')
btc
```

```python
#Now ethereum
eth = pd.DataFrame(columns = ['Date', 'Tweet', 'Likes', 'Retweets'])

for i in range(len(tweets)):
    cur = tweets.loc[i].Tweet
    cur = cur.lower()
    if (' eth ' in cur) | ('ethereum' in cur):
        eth = eth.append({'Date' : tweets.loc[i].Date, 'Tweet' : tweets.loc[i].Tweet, 'Likes' : tweets.loc[i].Likes, 'Retweets' : tweets.loc[i].Retweets}, 
                ignore_index = True)

eth['Date'] = pd.to_datetime(eth['Date'], format='%Y-%m-%d')
eth
```

```python
#Now crypto
crypto = pd.DataFrame(columns = ['Date', 'Tweet', 'Likes', 'Retweets'])

for i in range(len(tweets)):
    cur = tweets.loc[i].Tweet
    cur = cur.lower()
    if 'crypto' in cur:
        crypto = crypto.append({'Date' : tweets.loc[i].Date, 'Tweet' : tweets.loc[i].Tweet, 'Likes' : tweets.loc[i].Likes, 'Retweets' : tweets.loc[i].Retweets}, 
                ignore_index = True)

crypto['Date'] = pd.to_datetime(crypto['Date'], format='%Y-%m-%d')
crypto
```

```python
labels = ["Very Low", "Low", "Moderate", "High", "Very High"]
btc['Reach'] = pd.cut(btc['Retweets'],[0,8000,16000,60000,120000,100000000000], labels = labels)

eth['Reach'] = pd.cut(eth['Retweets'],[0,8000,16000,60000,120000,100000000000], labels = labels)

doge['Reach'] = pd.cut(doge['Retweets'],[0,8000,16000,60000,120000,100000000000], labels = labels)
```
