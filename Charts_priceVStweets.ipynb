```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#BTC historic data
BTC_USD = pd.read_csv('BTC-USD.csv')
BTC_USD['Date'] = pd.to_datetime(BTC_USD['Date'], format='%Y-%m-%d')
#BTC tweets
btc = pd.read_csv('btc_tweets.csv')
btc['Date'] = pd.to_datetime(btc['Date'], format='%Y-%m-%d')
```

```python
#plot chart price hist with tweet dates
plt.figure(figsize=(15, 7))
ax = sns.lineplot(x = BTC_USD.Date[BTC_USD.Date > '2017-10-01'], y = BTC_USD.Close)

for i in btc.Date:
    x = pd.Timestamp(i)
    plt.axvline(x,color='r', linewidth = 0.5)


ax.set_title("Bitcoin Price")
```

```python
doge_tweets_price = pd.read_csv('doge_tweets_price.csv')
btc_tweets_price = pd.read_csv('btc_tweets_price.csv')
eth_tweets_price = pd.read_csv('eth_tweets_price.csv')
```

```python
round((doge_tweets_price['tweet_day_two_price'][0] - doge_tweets_price['tweet_day_minustwo_price'][0])/doge_tweets_price['tweet_day_minustwo_price'][0],3)*100
```

```python
#tendency for tweet date -2 to tweet date +2
doge_tweets_price['growth_tend_prior'] = round((doge_tweets_price['tweet_day_price'] - doge_tweets_price['tweet_day_minustwo_price'])/doge_tweets_price['tweet_day_minustwo_price'],3)*100
doge_tweets_price['growth_tend_5days'] = round((doge_tweets_price['tweet_day_two_price'] - doge_tweets_price['tweet_day_minustwo_price'])/doge_tweets_price['tweet_day_minustwo_price'],3)*100
doge_tweets_price['growth_1day'] = round((doge_tweets_price['tweet_day_one_price'] - doge_tweets_price['tweet_day_price'])/doge_tweets_price['tweet_day_price'],3)*100
doge_tweets_price['growth_2day'] = round((doge_tweets_price['tweet_day_two_price'] - doge_tweets_price['tweet_day_price'])/doge_tweets_price['tweet_day_price'],3)*100
doge_tweets_price['growth_3day'] = round((doge_tweets_price['tweet_day_three_price'] - doge_tweets_price['tweet_day_price'])/doge_tweets_price['tweet_day_price'],3)*100

btc_tweets_price['growth_tend_prior'] = round((btc_tweets_price['tweet_day_price'] - btc_tweets_price['tweet_day_minustwo_price'])/btc_tweets_price['tweet_day_minustwo_price'],3)*100
btc_tweets_price['growth_tend_5days'] = round((btc_tweets_price['tweet_day_two_price'] - btc_tweets_price['tweet_day_minustwo_price'])/btc_tweets_price['tweet_day_minustwo_price'],3)*100
btc_tweets_price['growth_1day'] = round((btc_tweets_price['tweet_day_one_price'] - btc_tweets_price['tweet_day_price'])/btc_tweets_price['tweet_day_price'],3)*100
btc_tweets_price['growth_2day'] = round((btc_tweets_price['tweet_day_two_price'] - btc_tweets_price['tweet_day_price'])/btc_tweets_price['tweet_day_price'],3)*100
btc_tweets_price['growth_3day'] = round((btc_tweets_price['tweet_day_three_price'] - btc_tweets_price['tweet_day_price'])/btc_tweets_price['tweet_day_price'],3)*100

eth_tweets_price['growth_tend_prior'] = round((eth_tweets_price['tweet_day_price'] - eth_tweets_price['tweet_day_minustwo_price'])/eth_tweets_price['tweet_day_minustwo_price'],3)*100
eth_tweets_price['growth_tend_5days'] = round((eth_tweets_price['tweet_day_two_price'] - eth_tweets_price['tweet_day_minustwo_price'])/eth_tweets_price['tweet_day_minustwo_price'],3)*100
eth_tweets_price['growth_1day'] = round((eth_tweets_price['tweet_day_one_price'] - eth_tweets_price['tweet_day_price'])/eth_tweets_price['tweet_day_price'],3)*100
eth_tweets_price['growth_2day'] = round((eth_tweets_price['tweet_day_two_price'] - eth_tweets_price['tweet_day_price'])/eth_tweets_price['tweet_day_price'],3)*100
eth_tweets_price['growth_3day'] = round((eth_tweets_price['tweet_day_three_price'] - eth_tweets_price['tweet_day_price'])/eth_tweets_price['tweet_day_price'],3)*100
```

```python
#tendency positive or negative?
eth_tweets_price['label_tend_5days'] = eth_tweets_price['growth_tend_5days'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
btc_tweets_price['label_tend_5days'] = btc_tweets_price['growth_tend_5days'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
doge_tweets_price['label_tend_5days'] = doge_tweets_price['growth_tend_5days'].apply(lambda x: 'Positive' if x>=0 else 'Negative')

eth_tweets_price['label_tend_prior'] = eth_tweets_price['growth_tend_prior'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
btc_tweets_price['label_tend_prior'] = btc_tweets_price['growth_tend_prior'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
doge_tweets_price['label_tend_prior'] = doge_tweets_price['growth_tend_prior'].apply(lambda x: 'Positive' if x>=0 else 'Negative')

eth_tweets_price['label_tend_after'] = eth_tweets_price['growth_2day'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
btc_tweets_price['label_tend_after'] = btc_tweets_price['growth_2day'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
doge_tweets_price['label_tend_after'] = doge_tweets_price['growth_2day'].apply(lambda x: 'Positive' if x>=0 else 'Negative')
```

```python
labels = ["Very Low", "Low", "Moderate", "High", "Very High"]
btc_tweets_price['Reach'] = pd.cut(btc_tweets_price['Retweets'],[0,8000,16000,60000,120000,100000000000], labels = labels)

eth_tweets_price['Reach'] = pd.cut(eth_tweets_price['Retweets'],[0,8000,16000,60000,120000,100000000000], labels = labels)

doge_tweets_price['Reach'] = pd.cut(doge_tweets_price['Retweets'],[0,8000,16000,60000,120000,100000000000], labels = labels)
```
